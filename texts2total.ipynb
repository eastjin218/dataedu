```python
texts = glob.glob('../stt 정리/*.txt')
for te in texts:
    with open(te,'r',encoding= 'utf-8') as f:
        tmp = f.readlines()
        for i in tmp:
            if i == '\n':
                pass
            else:
                with open('../1_total.txt','a', encoding='utf-8') as f:
                    f.write(i)            
```
